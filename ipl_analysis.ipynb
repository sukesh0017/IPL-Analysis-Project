```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
matches =pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
matches.head()
```

```python
matches['winner'].value_counts().plot(kind='bar')
plt.title("Most Successful IPL Teams")
plt.show()
```

```python
top_scorers = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

top_scorers.plot(kind='bar')
plt.title("Top Run Scorers")
plt.show()
```

```python
deliveries['bowler'].value_counts().head(10).plot(kind='bar')
plt.title("Top Wicket Takers")
plt.show()
```

```python
matches['venue'].value_counts().head(10).plot(kind='bar')
plt.title("Top Venues")
plt.show()
```
